```python
#1 Calcular el rendimiento y el riesgo S&P 500

Stock_0 = DataFromSource('.SPX', interval_period="1D", initial = 365*10)
Precios = Stock_0.get_prices()
Rendimientos = Stock_0.get_returns()
```

```python
Precios[f'SMA_200'] = ta.SMA(Precios['.SPX'], timeperiod=200)
Precios['bear_market'] = Precios['.SPX'] < Precios[f'SMA_200']
Precios = Precios.reset_index()
Precios['year_month'] = Precios['Date'].dt.to_period('M')
max_day_indices = Precios.groupby('year_month')['Date'].idxmax()
Precios_max_day = Precios.loc[max_day_indices]
Precios_max_day = Precios_max_day.dropna()
Precios_max_day.set_index('Date', inplace = True)
Precios_max_day
```

```python
spx = ld.get_data('0#.SPX', fields=['TR.CommonName', 'TR.PriceClose',
                                      'TR.Volume', 'TR.TotalReturnYTD'])
spx
```

```python
spx_50 = spx.sample(n=50, random_state=42)
spx_50['random_numbers'] = np.random.rand(len(spx_50))
column_sum = spx_50['random_numbers'].sum()
spx_50['portfolio_share'] = spx_50['random_numbers'] / column_sum
spx_50
```
